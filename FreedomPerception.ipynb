```python
# Initial data exploration
display(df.shape, "", df.dtypes, "", df.isnull().sum(), "", df.describe())
```

```python
#Drop irrelevant features
df = df.drop(columns=['GINI index (World Bank estimate)', 'GINI index (World Bank estimate), average 2000-15'], axis=1)
```

```python
#Replace NaN values with mean or median depending on skewness
from scipy.stats import skew

na_columns = [col for col in df.columns if df[col].isnull().any()]

for col in na_columns:
    skew_value = skew(df[col].dropna())
    
    if abs(skew_value) < 1:
        df[col] = df[col].fillna(df[col].mean())#Normal distribution
    else:
        df[col] = df[col].fillna(df[col].median())#Skewed distribution
```

```python
#Spot outliers
columns = df.columns.tolist()
del columns[0]

for col in columns:
    sns.boxplot(x = df[col])
    plt.title(col)
    plt.show()
```

```python
#winsorize dataset by country
from scipy.stats.mstats import winsorize

cols_to_winsorize = [
    'Social support', 
    'Healthy life expectancy at birth', 
    'Freedom to make life choices', 
    'Generosity', 
    'Perceptions of corruption', 
    'Negative affect', 
    'Confidence in national government', 
    'Standard deviation of ladder by country-year', 
    'Standard deviation/Mean of ladder by country-year', 
    'gini of household income reported in Gallup, by wp5-year'
]

limits = (0.05, 0.05)
countries = df['country'].unique().tolist()

for country in countries:# I am winsorizing each feature per grouped-examples by country
    country_df = df[df['country'] == country].copy()
    
    for col in cols_to_winsorize:
        country_df[col] = winsorize(country_df[col], limits=limits).data
        
    df.loc[country_df.index, cols_to_winsorize] = country_df[cols_to_winsorize]
```

```python
#Check to see if outliers have been reduced
columns = df.columns.tolist()
del columns[0]

plt.clf()
for col in columns:
    sns.boxplot(x = df[col])
    plt.title(col)
    plt.show()
```

```python
#Drop country column from dataframe
#I decided to not one-hot encode cause it will be 164 new features, too few examples
#I also believe that winsorizing by country retained country-related aspects in the dataframe
df = df.drop(columns='country', axis=1)
```

```python
#Data model preproccessing
from sklearn.model_selection import train_test_split, GridSearchCV

#Models
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

#Evaluation
from sklearn.metrics import r2_score, mean_squared_error
```

```python
#Create Labeled Examples
X = df.drop(columns='Freedom to make life choices', axis=1)
y = df['Freedom to make life choices']
```

```python
#Create training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.30, random_state=1234)
```

<h4><b>LinearRegression model</b></h4>

```python
#Train
lr_model = LinearRegression().fit(X_train, y_train)
```

```python
#Test
y_lr_pred = lr_model.predict(X_test)
```

```python
#Evaluate

#RMSE
lr_rmse = mean_squared_error(y_test, y_lr_pred, squared=False)

#R2
lr_r2 = r2_score(y_test, y_lr_pred)
print(f'[LR] Root Mean Squared Error: {lr_rmse}')#Lower is better
print(f'[LR] R2: {lr_r2}')#Higher is better
```

<h4><b>RandomForestRegressor model</b></h4>

```python
param_grid = {'n_estimators':[30, 50, 70], 'max_depth':[6, 9, 12], 'min_samples_leaf':[10, 20, 30]}
```

```python
#Gridsearch to find optimal parameters
print('Running Grid Search...')

rf_regressor = RandomForestRegressor()
rf_grid = GridSearchCV(rf_regressor, param_grid, cv=5, scoring='neg_root_mean_squared_error')
rf_grid_search = rf_grid.fit(X_train, y_train)

print('Done')
```

```python
#RMSE
rmse_RF = -1 * rf_grid_search.best_score_
print(f"[DT] RMSE for the best model is {rmse_RF}")
```

```python
#Best parameters
rf_best_params = rf_grid_search.best_estimator_
rf_best_params
```

```python
#Create new model to train using best params
rf_model = RandomForestRegressor(max_depth=9, min_samples_leaf=10, n_estimators=70)
rf_model.fit(X_train, y_train)
```

```python
#Test
y_rf_pred = rf_model.predict(X_test)

#RMSE
rf_rmse = mean_squared_error(y_test, y_rf_pred, squared=False)

#r2
rf_r2 = r2_score(y_test, y_rf_pred)

print(f'[RF] Root Mean Squared Error: {rf_rmse}')#Lower is better
print(f'[RF] R2: {rf_r2}')#Higher is better
```

<h4><b>Visualize and Compare Model Performance</h4></b>

```python
RMSE_Results = [lr_rmse, rf_rmse]
R2_Results = [lr_r2, rf_r2]

rg = np.arange(2)
width = 0.35

#Bar plot with RMSE results
plt.bar(rg, RMSE_Results, width=width, label='RMSE')

#Bar plot with R2 results
plt.bar(rg+width, R2_Results, width=width, label='R2')

labels = ['LinearRegression', 'RandomForest']
plt.xticks(rg+width/2, labels)

plt.xlabel('Models')
plt.ylabel('RMSE/R2')

plt.ylim([0.05,.7])#I used this instead of [0, 1] to better show the difference in model performance
plt.title('Model Performance')
plt.legend(loc='upper left', ncol=2)
plt.show()

```

```python
#Quick bias/variability checker

#Linear` Regression
check_lr_pred = lr_model.predict(X_train)
check_rmse = mean_squared_error(y_train, check_lr_pred, squared=False)
check_r2 = r2_score(y_train, check_lr_pred)
print(f'[LR] Root Mean Squared Error: {lr_rmse}')#Lower is better
print(f'[LR] R2: {lr_r2}')#Higher is better
print(f'[LR_Check(Train)] Root Mean Squared Error: {check_rmse}')#Lower is better
print(f'[LR_Check(Train)] R2: {check_r2}')#Higher is better
print("")

#Random Forest Regression
check_rf_pred = rf_model.predict(X_train)
check1_rmse = mean_squared_error(y_train, check_rf_pred, squared=False)
check1_r2 = r2_score(y_train, check_rf_pred)
print(f'[RF] Root Mean Squared Error: {rf_rmse}')#Lower is better
print(f'[RF] R2: {rf_r2}')#Higher is better
print(f'[RF_Check(Train)] Root Mean Squared Error: {check1_rmse}')#Lower is better
print(f'[RF_Check(Train)] R2: {check1_r2}')#Higher is better
```
